```python
from robustica import RobustICA, sampledata
X = sampledata
```

## with Icasso method

```python
rica = RobustICA(n_components=10,
                 n_jobs=10,
                 robust_method='icasso', 
                 robust_kws={'distance_threshold':0.8,'n_clusters':None})
rica.fit(X.values)
```

```python
rica.clustering_stats_
```

## with Sastry method

```python
rica = RobustICA(n_components=10,
                 n_jobs=10,
                 robust_method='sastry', 
                 robust_kws={'min_samples':5, 'n_jobs':10})
rica.fit(X)
```

```python
rica.clustering_stats_
```

## with DBSCAN

```python
rica = RobustICA(n_components=10,
                 n_jobs=10,
                 robust_method='DBSCAN', 
                 robust_kws={'min_samples':5, 'n_jobs':10})
rica.fit(X.values)
```

```python
rica.clustering_stats_
```
